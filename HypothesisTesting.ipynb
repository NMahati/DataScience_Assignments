```python
import math
x_bar = 3050.0        # Sample mean weekly cost (Rs.)
X_mean = 600.0         # Mean number of units produced
X_sd = 25.0            # Standard deviation of units
n = 25                 # Sample size
alpha = 0.05
```

Hypothesis Statement

- H0: μ = 1000 + 5*600 = 4000  (Weekly cost follows model)
- H1: μ > 4000  (Weekly cost is higher than model suggests)

```python
mu = 1000 + 5 * X_mean
```

```python
sigma_w = 5 * X_sd
```

```python
# Standard error of mean
se = sigma_w / math.sqrt(n)
print(se)
```

```python
z_stat = (x_bar - mu) / se
print(z_stat)
```

```python
#One-tailed test (right side), α = 0.05
z_critical = 1.6448536269514722 
```

```python
if z_stat > z_critical:
    decision = "Reject H0"
else:
    decision = "Fail to Reject H0"
```

```python
if z_stat > z_critical:
    conclusion = ("There is sufficient evidence at α = 0.05 "
                  "to conclude weekly operating costs are higher than the model suggests.")
else:
    conclusion = ("There is NOT sufficient evidence at α = 0.05 "
                  "to conclude weekly operating costs are higher than the model suggests.\n"
                  "In fact, the sample mean is lower than the model's expected cost.")
```

```python
print(f"1. Critical value (Zα)   = {z_critical:.4f}")
print(f"2. Decision: {decision}")
print(f"3. Conclusion:")
print(f"   {conclusion}")
```
