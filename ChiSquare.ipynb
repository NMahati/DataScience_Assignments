```python
import numpy as np
from scipy.stats import chi2
```

```python
observed = np.array([
    [50, 70],   # Very Satisfied
    [80, 100],  # Satisfied
    [60, 90],   # Neutral
    [30, 50],   # Unsatisfied
    [20, 50]    # Very Unsatisfied
])
```

```python
row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
grand_total = observed.sum()
```

- H0: There is NO association between device type and customer satisfaction.
- H1: There IS a significant association between device type and customer satisfaction.

```python
expected = np.outer(row_totals, col_totals) / grand_total
```

```python
chi_square_stat = ((observed - expected) ** 2 / expected).sum()
```

```python
rows, cols = observed.shape
df = (rows - 1) * (cols - 1)
```

```python
alpha = 0.05
chi_critical = chi2.ppf(1 - alpha, df)
```

```python
if chi_square_stat > chi_critical:
    decision = "Reject H0"
    conclusion = (
        "There is a significant association between device type and customer satisfaction."
    )
else:
    decision = "Fail to Reject H0"
    conclusion = (
        "There is NO significant association between device type and customer satisfaction."
    )
```

```python
print("1. Observed Frequencies:\n", observed)
print("\n2.Expected Frequencies (calculated):\n", np.round(expected, 2))

print("\n3. Chi-Square Statistic:")
print(f"   χ² = {chi_square_stat:.4f}")

print("\n4. Degrees of Freedom:")
print(f"   df = ({rows}-1) * ({cols}-1) = {df}")

print("\n5. Critical Value (α = 0.05):")
print(f"   χ²_critical = {chi_critical:.4f}")

print("\n6. Decision Rule:")
print("   Reject H0 if χ²_calculated > χ²_critical")

print(f"\n7. Comparison: χ²_calculated = {chi_square_stat:.4f}, χ²_critical = {chi_critical:.4f}")
print(f"   Decision: {decision}\n")

print("8. Conclusion:")
print("   " + conclusion)
```
