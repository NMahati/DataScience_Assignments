```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, accuracy_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input
from tensorflow.keras.optimizers import Adam
```

```python
X = data.drop(columns=['letter'])   # features (numeric only)
y = data['letter']                  # target (alphabets)
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42
)
```

```python
model = Sequential()
```

```python
model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
```

```python
model.add(Dense(32, activation='relu'))
```

```python
model.add(Dense(len(np.unique(y_encoded)), activation='softmax'))
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    X_train, y_train,
    epochs=30,
    batch_size=32,
    validation_split=0.1,
    verbose=1
)
```

```python
input_dim = X_train.shape[1]
num_classes = len(np.unique(y_encoded))
```

```python
def create_model(input_dim, num_classes,
                 hidden_layers, neurons,
                 learning_rate, activation):

    # вњ… safety check INSIDE the function
    hidden_layers = max(1, hidden_layers)

    model = Sequential()
    model.add(Input(shape=(input_dim,)))

    for _ in range(hidden_layers):
        model.add(Dense(neurons, activation=activation))

    model.add(Dense(num_classes, activation='softmax'))

    optimizer = Adam(learning_rate=learning_rate)
    model.compile(
        optimizer=optimizer,
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
```

```python
param_grid = {
    "hidden_layers": [1, 2, 3],
    "neurons": [32, 64],
    "learning_rate": [0.001, 0.0005],
    "activation": ["relu", "tanh"]
}
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)
```

```python
best_accuracy = 0
best_params = {}
```

```python
for hl in param_grid["hidden_layers"]:
    for n in param_grid["neurons"]:
        for lr in param_grid["learning_rate"]:
            for act in param_grid["activation"]:

                temp_model = create_model(
                    input_dim=input_dim,
                    num_classes=num_classes,
                    hidden_layers=hl,
                    neurons=n,
                    learning_rate=lr,
                    activation=act
                )

                temp_model.fit(
                    X_train, y_train,
                    epochs=20,
                    batch_size=32,
                    validation_split=0.2,
                    callbacks=[early_stop],
                    verbose=0
                )

                loss, acc = temp_model.evaluate(
                    X_test, y_test, verbose=0
                )

                if acc > best_accuracy:
                    best_accuracy = acc
                    best_params = {
                        "hidden_layers": hl,
                        "neurons": n,
                        "learning_rate": lr,
                        "activation": act
                    }
```

```python
print("Best Accuracy:", best_accuracy)
print("Best Parameters:", best_params)
```

```python
final_model = create_model(
    input_dim=input_dim,
    num_classes=num_classes,
    hidden_layers=best_params["hidden_layers"],
    neurons=best_params["neurons"],
    learning_rate=best_params["learning_rate"],
    activation=best_params["activation"]
)
```

```python
final_model.fit(
    X_train,
    y_train,
    epochs=30,
    batch_size=32,
    verbose=1
)
```

```python
y_pred = np.argmax(final_model.predict(X_test), axis=1)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))
```
